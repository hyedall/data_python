### 데이터프레임 생성
---

````python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(10, 26).reshape(4, 4,), 
                  columns = ['a', 'b', 'c' ,'d'])
df
````




### 데이터프레임 인덱싱
___
1. 열 기준 인덱싱
- 하나의 열 : 시리즈 형태로 가져옴
```python
df[열이름] 
```
&nbsp;
- 두개 이상 : 데이터 프레임으로 가져옴  
(컬럼명을 리스트 형태로 넣기)
```python
df[[열이름1, 열이름2]] 
```
&nbsp;
- 열에 값 대입
```python
df['열이름'] = 값 # 열 전체가 한 값으로 대입됨 
df['열이름'] = 값1, 값2, 값3
df['열이름'] = np.arange(5000, 8900, 1000) # 5000부터 8900까지 1000씩 증가 
```
&nbsp;
2. 행 기준 인덱싱
- 슬라이싱 가능 
```python
df[1:3]
df['행1':'행3'] 
```
- 개별 요소 추출  
행 이름으로 가져오게 되서 주의 필요 !! (위치로 x) 
```python
df['열이름']['행이름'] 
```
&nbsp;
3. 인덱서
- loc 인덱서   df.loc['행이름', '열이름']
  - 행 이름으로 데이터 추출, 행이름으로 슬라이싱도 가능
  - 행 이름으로 여러행 추출 (데이터프레임으로) 
  ```python
df.loc[['행이름1', '행이름2']]
```
  - 인덱스 값으로 조건식 사용 가능

    ```python
    df.loc[df.열이름 > 조건] # 인덱스 값으로 조건식을 줄 수 있음 
    # 인덱스 값을 반환하는 함수의 결과값을 사용
    def sel_row(df):
      return df.열이름 > 조건식
    df.loc[sel_row(df)]
    ```

  - df.loc['행', '열']

### 데이터프레임 특정 조건 데이터 필터링
___


```python
condition = (df.열이름1 == 조건1) & (df.열이름2 == 조건2) # 조건식 작성

df[condition]
```
```python
df[df.열이름 == 조건]
```

```python
df.loc[df.열이름 > 조건] # 인덱스 값으로 조건식을 줄 수 있음 

# 인덱스 값을 반환하는 함수의 결과값을 사용
def sel_row(df):
  return df.열이름 > 조건식
df.loc[sel_row(df)]
```
